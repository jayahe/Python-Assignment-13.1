```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from pandasql import sqldf
```

```python
url1 = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
url2 = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names'
cols = ["age", "type_employer", "fnlwgt", "education", 
                "education_num","marital", "occupation", "relationship", "race","sex",
                "capital_gain", "capital_loss", "hr_per_week","country", "income"]
df = pd.read_csv(url1,sep = ',',names = cols)
df.head()
```

```python
# ?pd.read_csv()
pysqldf = lambda q: sqldf(q,globals())
df.info()
```

### 1. Select 10 records from the adult sqladb

```python
q = '''Select * from df limit 10'''
print(q)
pysqldf(q)
```

### 2. Show me the average hours per week of all men who are working in private sector

```python
df.type_employer.unique()
```

```python
q = '''Select   sex
              , type_employer
              , cast(round(avg(hr_per_week)) as int) as avg_hrperweek
             from 
                df
              where 
                sex = ' Male' and type_employer = ' Private'
              group by 
                  sex
                , type_employer;
              '''
pysqldf(q)
```

### 3. Show me the frequency table for education, occupation and relationship, separately

```python
df.shape
```

```python
q1 = 'Select education, count(education) as Count from df group by education'
q2 = 'Select occupation, count(occupation) as Count from df group by occupation'
q3 = 'Select relationship, count(relationship) as Count from df group by relationship'

print('Eduction Frequency: \n',pysqldf(q1))
print('Occupation Frequency: \n',pysqldf(q2))
print('Relationship Frequency: \n',pysqldf(q3))
```

### 4. Are there any people who are married, working in private sector and having a masters degree

```python
q = '''Select marital, 
              type_employer,
              education,
              count(marital) as Count 
              from 
                df 
              where
                marital = ' Married-civ-spouse' and
                type_employer = ' Private' and
                education = ' Masters'
              group by
                marital, type_employer, education;
              '''
pysqldf(q)
```

### 5. What is the average, minimum and maximum age group for people working in different sectors

```python
q = '''Select occupation,
              min(age) as min_age,
              max(age) as max_age,
              cast(round(avg(age)) as int) as avg_age
              from
                df
              group by
                type_employer;
                '''
print(q)
pysqldf(q)
```

### 6. Calculate age distribution by country

```python
q = '''Select country
          , cast(round(avg(age)) as int) as Age
            from 
              df
            group by 
               country;
            '''
print(q)
pysqldf(q)
```

### 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
q = '''Select *,
              ([capital_gain] -[capital_loss]) as Net_Capital_Gain
              from
                df;
            '''
print(q)
pysqldf(q).tail()
```
